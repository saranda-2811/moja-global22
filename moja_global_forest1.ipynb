The Congo Basin is the second largest forest in the world, and the largest rainforest in Africa. It spans mostly across 6 countries – 
1.	Cameroon
2.	The Republic of the Congo
3.	The Central African Republic
4.	The Democratic Republic of the Congo
5.	Equatorial Guinea
6.	Gabon

These dense old-growth forests have huge volumes of carbon in their vegetation and tree trunks(39 billion tonnes), which serve as an important buffer against climate change. https://royalsocietypublishing.org/doi/10.1098/rstb.2012.0376
African forests contain about 25% of tropical forest carbon.

The forests of the Democratic Republic of Congo store the most carbon of any African country, accounting for 38-56% of carbon. Three other Congo Basin countries — Cameroon, Republic of Congo, and Central African Republic — follow.

https://news.mongabay.com/2011/06/african-forests-store-25-of-tropical-forest-carbon/


It has very rich biodiversity – including more than 600 tree species, and 10,000 animal species. 

If we take the global ecological zones data, we will see it has only one zone listed in the Congo Basin - Tropical Dry Forest.
Hence, we have to list GEZs separately according to the countries spanning the Basin. 

Then, the Congo Basin is classified into 6 ecological zones:
1.	Tropical dry forest
2.	Tropical moist forest
3.	Tropical mountain system
4.	Tropical rainforest
5.	Tropical shrubland
6.	Water

Types of climates in the basin:
1.	Boreal dry
2.	Boreal moist
3.	Polar dry
4.	Polar moist
5.	Tropical moist
6.	Tropical montane
7.	Tropical wet

Types of agro-ecological zones:
1.	Dry – moderate, poor soils
2.	Humid – good, moderate, poor soils
3.	Hydromorphic soils
4.	Irrigated soils
5.	Moist – good, moderate, poor soils
6.	Steep Terrain
7.	Sub-humid – good, moderate, poor soils
8.	Water

Central Africa's deforestation rate between 1990-2010 was the lowest of any major forest region in the world. However deforestation trended upward during the 2010s with expansion of industrial logging and conversion for large-scale agriculture. The biggest threats to the Congo rainforest come from industrial plantations, especially for palm oil, rubber, and sugar production.

The Democratic Republic of Congo contains about 60% of the Basin, but it's forests are facing significant threats from firewood collection, logging, mining, and slash and burn agriculture, as well as the impacts of poaching.

```python
def getdf(link):
  req = requests.get(link).content
  return gpd.read_file(io.StringIO(req.decode('utf-8')))
```

## 2. CLIMATE

```python
climate = 'https://datasets.mojaglobal.workers.dev/0:/Climate/IPCC_ClimateZoneMap_Vector.geojson'
climate_df = getdf(climate)
```

```python
classification =  {'1': 'Tropical montane', '2' : 'Tropical wet', 
                   '3': 'Tropical moist', '4': 'Tropical dry', 
                   '5': 'Warm temperate moist', '6': 'Warm temperate dry',
                   '7': 'Cool temperate moist', '8': 'Cool temperate dry',
                   '9': 'Boreal moist', '10': 'Boreal dry', '11' : 'Polar moist',
                   '12': 'Polar dry'}

# normalize column's name
climate_df.rename(columns={'CLASS_NAME': 'class_name'}, inplace=True)

# remove oceans
climate_df = climate_df[climate_df.class_name != '0']

# replace classification codes with actual names
climate_df.class_name = climate_df.class_name.astype('str').replace(classification)
 
# get only necessary data related to forest
forest_climate_df = climate_df.clip(congo_countries_df)
```

```python
forest_climate_df.plot(column='class_name', linewidth=3, figsize=(15,10), legend=True)
```

## 3. BIOCLIMATIC AND ECOLOGICAL ZONES

3.1 AGRO-ECOLOGICAL ZONES

```python
agro_eco_zones = 'https://datasets.mojaglobal.workers.dev/0:/Bioclimatic&EcologicalZones/GAEZ_Global_AgroEcological_Zones/GlobalAgroEcologicalZones_GAEZ.geojson'
agroeco_df = getdf(agro_eco_zones)
```

```python
values = {'1': 'Steep terrain', '2': 'Artic/Cold', '3': 'Desert/Arid', 
          '4': 'Irrigated Soils', '5': 'Hydromorphic soils', 
          '6': 'Dry, good soils', '7': 'Dry, moderate soils', '8': 'Dry, poor soils', 
          '9': 'Moist, good soils', '10': 'Moist, moderate soils', 
          '11': 'Moist, poor soils', '12': 'Sub-humid, good soils', 
          '13': 'Sub-humid, moderate soils', '14': 'Sub-humid, poor soils', 
          '15': 'Humid, good soils', '16': 'Humid, moderate soils', 
          '17': 'Humid, poor soils', '18': 'Water'}

# normalize column's name
agroeco_df.rename(columns={'Value': 'value'}, inplace=True) 

# replace classification codes with actual names
agroeco_df.value = agroeco_df.value.astype('str').replace(values)
```

```python
congo_agroeco_zones_df = agroeco_df.clip(congo_countries_df)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
ax.set_title('AGRO ECOLOGICAL ZONES OF THE CONGO BASIN')

congo_agroeco_zones_df.plot(ax=ax, column='value', linewidth=3, legend=True)
plt.show()
```

3.3 HOLDRIDGE LIFE ZONES

The Holdridge life zones system is a global bioclimatic scheme for the classification of land areas. Following are the types of holdridge life zones in the Congo Basin

```python
holdridge_zones = 'https://datasets.mojaglobal.workers.dev/0:/Bioclimatic&EcologicalZones/HoldridgeLifeZones.json'
holdridge_zones_df = getdf(holdridge_zones)
```

```python
# normalize column's name
holdridge_zones_df.rename(columns={'DESC': 'desc'}, inplace=True)
```

```python
# get only necessary data related to forest
holdridge_zones_df = holdridge_zones_df[['desc', 'geometry']]
congo_holdridge_zones_df = holdridge_zones_df.clip(congo_countries_df)
```

```python
# visualize ecological zones on the map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

ax.set_title('HOLDRIDGE LIFE ZONES OF CONGO BASIN')

congo_holdridge_zones_df.plot(ax=ax, column='desc', legend=True)
plot_countries(ax)
plt.show()
```
